#### Plotting helper function


A function to plot histograms with distribution curves for the **feature** passed as an argument. The number of **bins** is decided based on the data. Each subgroup of features are organized using **ax** for each group of features we use subplot to organize them together.

As a rule of thumb when using subplots, the size will be (5\*columns,3.3\*rows).

Number of rows are 3 max and colums will depend on the number of features

````python
def plot_hist(feature, bin, ax):
  """
  Plot histogram for the given feature
  """
  sns.histplot(df[feature], bins=bin, kde=True, ax=ax) #kde to plot distribution curve
  ax.set_title(f'{feature} Distribution') #Label each plot


````

## **Normaility Testing**

#### Normaility testing helper function

To start testing we first state out null hypothesis, $H_o$ , that the features are normally distributed and the alternate hypothesis, $H_1$ that they are not noramlly distributed.

We make a function that uses Shapiro-Wilk test whcih either, rejects our hypothesis or fails to(meaning it likely stands)



````python
from scipy.stats import shapiro

def normality_test(x , alpha=0.05):
   # Perform Shapiro-Wilk test
    W_statistic, p_value = shapiro(x)

    # Determine if data is normal
    is_normal = (p_value >= alpha) # If p>0.05 then H_o not rejected

    # Determine conclusion
    if is_normal:
        conclusion = "Fail to reject"
    else:
        conclusion = "Rejected"

    return conclusion
````

````python
x = [0,1,2,2,3,3,3,4,4,5,6]

hypo_table = pd.DataFrame(index=df.columns, columns=['H_o status'])

for col in df.columns:
  stat = normality_test(col)
  hypo_table.loc[col] = stat

# Create figure
fig, ax = plt.subplots(figsize=(8, 0.4 * len(hypo_table) + 1))
ax.axis('off') # more organized table

# Build table
table = ax.table(
    cellText=hypo_table.reset_index().values,
    colLabels=['Feature', 'H status'],
    cellLoc='center',
    loc='center'
)

# Formatting
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.4)

#To save an image use the next line
#plt.savefig("hypothesis_table.png", dpi=300, bbox_inches="tight")

plt.show()
````

#### Test Comment
The normality test seems accurate, considering that none of the data is normally distributed as shown by the previous plots.

## Correlation Coefficient

First we create a function that calculates the correlation coefficient value, r, using the following formula:

$$r= \frac{Cov(X,Y)}{_x  _y}$$

which is expanded as:

$$
r= \frac
{_{i=1}^n (x_i - \bar{x})  (y_i - \bar{y}) }
{\sqrt{_{i=1}^n (x_i - \bar{x})^2} \cdot \sqrt{_{i=1}^n (y_i - \bar{y})^2}}
$$

Where x is each predictor and y is the result "life expectency"

We will use the methods .mean() and .std(), so the final r formula is:

$$
r= \frac
{_{i=1}^n (x_i - \bar{x})  (y_i - \bar{y}) }
{(n-1) \cdot _x \cdot _y}
$$


````python
#Start with the y  & mean to avoid recalculations

mean_y = df['Life_expectancy'].mean()
std_y = df['Life_expectancy'].std()
n = len(df['Life_expectancy']) #number of samples

def corr(x):
  """
  A function to calculate the correlation coefficient r between
  the given feature and the result, life expectancy.
  """
  #for predictor x
  mean_x = df[x].mean() #x mean
  std_x = df[x].std()   # x standard deviation

  #For the numerator
  cov_xy = np.sum((df[x] - mean_x) * (df['Life_expectancy'] - mean_y))

  #For the denominator
  std_xy = (n-1) * std_x * std_y

  #Final r value
  r = cov_xy / std_xy

  return r
````

````python
#Create a list of a predictors names
predictor_col = df.columns.drop('Life_expectancy')

#A table for the r values
corr_table = pd.DataFrame(index=predictor_col, columns=['r'])

#Populate the table
for col in predictor_col:
  r = corr(col)
  corr_table.loc[col] = r
  #print(f'The correlation coefficient for {col} is {r}')

#sort the r values
corr_table = corr_table.sort_values(by='r', ascending=False)
corr_table['r'] = corr_table['r'].astype(float).round(4) # Found to 4 decimals



fig, ax = plt.subplots(figsize=(7, 0.4 * len(corr_table) + 1))
ax.axis('off')

# Build table
table = ax.table(
    cellText=corr_table.reset_index().values,
    colLabels=['Predictor', 'r'],
    cellLoc='center',
    loc='center'
)

# Formatting
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.4)

# To save the figure use the next line
plt.savefig("correlation_table.png", dpi=300, bbox_inches="tight")

plt.show()
````
